```python
# Q1. What is the average, median, and standard deviation of house prices?

average_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()

print("Average Price:", average_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_price)
```

## Explanation

- mean() -> average price
- median() -> middle value
- std() -> spread of prices

## Interpretation

- The average price represents the overall market value of houses in the dataset.
- The median price shows the middle house price and helps understand the central tendency without the effect of extreme values.
- The standard deviation indicates how much house prices vary from the average price.
- A high standard deviation means house prices differ greatly, while a low value means prices are more consistent.

```python
# Q2. Which number of bedrooms is most common?

common_bedrooms = df['number of bedrooms'].mode()[0]

print("Most Common Bedrooms:", common_bedrooms)
```

## Explanation

 - mode() gives the most frequent value.

 ## Interpretation

- The most common number of bedrooms represents the type of house most frequently available or preferred in the market.
- This helps identify common housing demand among buyers.
- If 3-bedroom houses are most common, it suggests medium-sized family homes are popular.

```python
# Q3. Check if the price data is skewed.

skewness = df['Price'].skew()

print("Skewness:", skewness)
```

## Expalantion

- skew() checks data distribution.

## Interpretation

- If the skewness value is positive, house prices are right-skewed, meaning a few expensive houses increase the average price.
- If skewness is negative, lower-priced houses dominate the dataset.
- House price datasets are usually positively skewed because luxury homes have extremely high prices compared to normal houses.

```python
# Q4. What is the average price for each number of bedrooms.

avg_price_bedroom = df.groupby('number of bedrooms')['Price'].mean()

print(avg_price_bedroom)
```

## Explanation

- Groups houses by bedroom count.
- Finds average price for each group.

## Interpretation

- This analysis helps understand how house prices change with bedroom count.
- Generally, houses with more bedrooms have higher prices because they provide more space and facilities.
- It helps buyers and real estate businesses understand pricing patterns based on house size.

```python
# Q5. What is the relationship between living area and and price? (Use : Area without Basement)

correlation = df['living area'].corr(df['Price'])

print("Correlation:", correlation)
```

## Explanation

- corr() measures relationship strength.

## Interpretation

- A positive correlation means larger living areas tend to have higher prices.
- A strong positive relationship indicates area is an important factor affecting house price.
- Buyers usually pay more for houses with larger living spaces.

```python
# Q6. Identify any anomalies where houses have high prices but low areas.

anomalies = df[(df['Price'] > df['Price'].quantile(0.90)) &
               (df['living area'] < df['living area'].quantile(0.25))]

print(anomalies[['Price', 'living area']])
```

## Explanation

Finds houses with:
- Very high price
- Small area

## Interpretation

- These houses are unusual because they have small areas but very high prices.
- Possible reasons may include:
(i) Premium location
(ii) Luxury interiors
(iii) Waterfront view
(iv) Special amenities
- Detecting anomalies helps identify unique or overpriced properties.

```python
# Q7. Compare average price based on number of floors and houses with or without waterfront.

# Average Price by Floors

floor_price = df.groupby('number of floors')['Price'].mean()

print(floor_price)
```

```python
# Average Price by Waterfront

waterfront_price = df.groupby('waterfront present')['Price'].mean()

print(waterfront_price)
```

## Explanation

- Comapres house prices using categories.

## Interpretaion for Floors

- Houses with more floors may have higher prices because they provide more usable space.
- Multi-floor houses are often considered more luxurious or suitable for large families.

## Interpretation for Waterfront

- Waterfront houses generally have higher prices because scenic locations increase property value.
- This shows location and surroundings strongly affect real estate prices.

```python
# Q8. Identify the minimum and maximum house price. What does this indicate?

minimum_price = df['Price'].min()
maximum_price = df['Price'].max()

print("Minimum Price:", minimum_price)
print("Maximum Price:", maximum_price)
```

## Explanation

- min() gives lowest price.
- max() gives highest price.

## Interpretation

- The minimum price represents the least expensive house in the dataset.
- The maximum price represents the most luxurious or premium property.
- The difference between minimum and maximum prices shows the price range and diversity of the housing market.

```python
# Q9. Which location (zipcode/area) has the highest average price?

highest_location = df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False)

print(highest_location.head(1))
```

## Explanation

- Groups prices by postal code.
- Finds highest average price location.

## Interpretation

- The location with the highest average price is likely a premium residential area.
- High prices may be due to:
(i) Better infrastructure
(ii) Good connectivity
(iii) Luxury facilities
(iv) High demand
- This analysis helps identify valuable real estate locations.
